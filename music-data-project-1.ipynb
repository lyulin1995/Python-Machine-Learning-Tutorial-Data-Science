```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
# split the dataset for training and testing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

music_data = pd.read_csv('music.csv')
music_data
```

```python
# preparing and cleaning the data 
# split the dataset: input dataset and output dataset 
X = music_data.drop(columns=['genre'])  # create a new dataset without genre column
# X  # the input dataset
y = music_data['genre']
# y # the output dataset: the prediction or the answers
# allocate 20% of data for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# create an instance of decision tree class
model = DecisionTreeClassifier()
# train the model
model.fit(X_train, y_train)
# ask the model to make prediction
predictions = model.predict(X_test)  # this method will pass 2D array
# predictions
```

```python
# calculating the accuracy: compare the prediction with the actually value we have
# expected values: y_test; actual values: predictions
score = accuracy_score(y_test, predictions) # range of 0 to 1
score
```
